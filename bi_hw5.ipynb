```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as sts
import math 
```

```python
df = pd.read_table('IB_statistics_data_HW5.txt',sep='\s+', engine='python')
```

##### Задача 2.а.

```python
a_2a = df.sort_index().sort_values('Ex2a', kind='mergesort')
res_2a = list(a_2a['Ex2a'])

F_t_2a = [res_2a[i] / 42 for i in range(len(res_2a))] #теореттическая функция распределения
F_e_2a = [0 for i in range(len(res_2a))]              #эмпирическая функция распределения

for i in range(len(res_2a)):
    j = 0
    while((res_2a[j] <= res_2a[i]) and j < len(res_2a) - 1):
        F_e_2a[i] += 1
        j += 1
    F_e_2a[i] /= len(res_2a)
    
D_n_2a = [abs(F_e_2a[i] - F_t_2a[i]) for i in range(len(res_2a))]
T_n_2a = math.sqrt(len(res_2a)) * max(D_n_2a)            #статистика критерия Колмогорова

lambda_2a = 1.36      #квантиль распределения Колмогорова для уровня значимости alpha = 0.05 взят из таблицы

if T_n_2a < lambda_2a:
    print('Принимаем гипотезу H_0')
else:
    print('Отвергаем гипотезу H_0')
    
h_2a = df['Ex2a'].hist()
fig = h_2a.get_figure()

fig_2 = plt.figure(figsize=(6, 6))
plt.plot(res_2a, F_t_2a)
plt.plot(res_2a, F_e_2a)

None
```

##### Задача 2.b.

```python
a_2b = df.sort_index().sort_values('Ex2b', kind='mergesort')
res_2b = list(a_2b['Ex2b'])

F_t_2b = [1 - math.exp(- 7 * res_2b[i]) for i in range(len(res_2b))] #теореттическая функция распределения
F_e_2b = [0 for i in range(len(res_2b))]                             #эмпирическая функция распределения

for i in range(len(res_2b)):
    j = 0
    while((res_2b[j] <= res_2b[i]) and j < len(res_2b) - 1):
        F_e_2b[i] += 1
        j += 1
    F_e_2b[i] /= len(res_2b)
    
    
D_n_2b = [abs(F_e_2b[i] - F_t_2b[i]) for i in range(len(res_2b))]
T_n_2b = math.sqrt(len(res_2b)) * max(D_n_2b)            #статистика критерия Колмогорова

lambda_2b = 1.36      #квантиль распределения Колмогорова для уровня значимости alpha = 0.05 взят из таблицы

if T_n_2b < lambda_2b:
    print('Принимаем гипотезу H_0')
else:
    print('Отвергаем гипотезу H_0')
    
h_2b = df['Ex2b'].hist()
fig = h_2b.get_figure()

fig_2 = plt.figure(figsize=(6, 6))
plt.plot(res_2b, F_t_2b)
plt.plot(res_2b, F_e_2b)

None
```

##### Задача 2.c.

```python
a_2c = df.sort_index().sort_values('Ex2c', kind='mergesort')
res_2c = list(a_2c['Ex2c'])

F_t_2c = [sts.norm.cdf(res_2c[i]) for i in range(len(res_2c))] #теореттическая функция распределения
F_e_2c = [0 for i in range(len(res_2c))]                       #эмпирическая функция распределения

for i in range(len(res_2c)):
    j = 0
    while((res_2c[j] <= res_2c[i]) and j < len(res_2c) - 1):
        F_e_2c[i] += 1
        j += 1
    F_e_2c[i] /= len(res_2c)
    
    
D_n_2c = [abs(F_e_2c[i] - F_t_2c[i]) for i in range(len(res_2c))]
T_n_2c = math.sqrt(len(res_2c)) * max(D_n_2c)            #статистика критерия Колмогорова

lambda_2c = 1.36      #квантиль распределения Колмогорова для уровня значимости alpha = 0.05 взят из таблицы

if T_n_2c < lambda_2c:
    print('Принимаем гипотезу H_0')
else:
    print('Отвергаем гипотезу H_0')
    
h_2c = df['Ex2c'].hist()
fig = h_2c.get_figure()

fig_2 = plt.figure(figsize=(6, 6))
plt.plot(res_2c, F_t_2c)
plt.plot(res_2c, F_e_2c)

None
```

##### Задача 3.а.

```python
res_3a = list(df['Ex3a'])

teta_3a = 0

for i in range(len(res_3a)):
    if res_3a[i] == 1:
        teta_3a += 1

teta_3a /= len(res_3a)

p_3a = 0.8
    
T_n_3a = math.sqrt(len(res_3a)) * (p_3a - teta_3a) / math.sqrt(p_3a * (1 - p_3a))  #статистика критерия  по общей схеме

h_l_3a = -1.96 #доверительная
h_u_3a = 1.96  #область

if T_n_3a < h_u_3a and T_n_3a > h_l_3a:
    print('Принимаем гипотезу H_0')
else:
    print('Отвергаем гипотезу H_0')
    
h_3a = df['Ex3a'].hist()
fig = h_3a.get_figure()
```
